## Section 1: Tools and Libraries

Before writing any code, it is good practice to list the tools and libraries this notebook will use. This helps the reader understand the dependencies upfront.

### Libraries Used

- `math` — for basic mathematical operations
- `random` — for generating random numbers
- `statistics` — for computing mean and standard deviation

### What This Notebook Will Do

2. Generate a sample list of numbers

## Section 2: Creating Sample Data

The next step is to generate a sample dataset. We will use `random.randint()` to produce a list of 10 integers between 1 and 100.

The function `random.randint(a, b)` returns a random integer N such that `a <= N <= b`.

> Run the cell below to generate the sample data.

````python
random.seed(42)
sample_data = [random.randint(1, 100) for _ in range(10)]

print("Sample Data:", sample_data)
````

The variable `sample_data` now holds a list of 10 random integers. The `random.seed(42)` call ensures the output is reproducible — running this cell multiple times will produce the same list.

---

## Section 3: Basic Statistical Analysis

Now that we have data, we will compute some basic descriptive statistics:

- **Mean** — the average value of the list, computed using `statistics.mean()`
- **Maximum** — the largest value, computed using the built-in `max()` function
- **Minimum** — the smallest value, computed using the built-in `min()` function
- **Square root of mean** — computed using `math.sqrt()`

> The cell below performs these calculations and prints the results.

````python
mean_value = statistics.mean(sample_data)
max_value = max(sample_data)
min_value = min(sample_data)
sqrt_mean = math.sqrt(mean_value)

print(f"Mean       : {mean_value}")
print(f"Maximum    : {max_value}")
print(f"Minimum    : {min_value}")
print(f"Sqrt(Mean) : {sqrt_mean:.4f}")
````

The output above shows four computed values from the sample dataset. The `:.4f` format specifier in the f-string rounds the square root to 4 decimal places for cleaner display.

## Section 4: Writing a Custom Function

Good notebooks document functions before they are defined. The function below, `describe_list()`, accepts a list of numbers and prints a short summary.

### Function Signature

```python
def describe_list(data: list) -> None:
    """
    Prints a summary of basic statistics for a numeric list.

    Parameters:
        data (list): A list of numeric values

    Returns:
        None
    """
```

The function uses `len()`, `min()`, `max()`, and `statistics.mean()` internally. No external libraries are needed beyond what was already imported in Section 1.

> The cell below defines the function and calls it with `sample_data`.

````python
def describe_list(data):
    """
    Prints a summary of basic statistics for a numeric list.

    Parameters:
        data (list): A list of numeric values

    Returns:
        None
    """
    print("--- List Summary ---")
    print(f"Count  : {len(data)}")
    print(f"Min    : {min(data)}")
    print(f"Max    : {max(data)}")
    print(f"Mean   : {statistics.mean(data)}")
    print("--------------------")

describe_list(sample_data)
````

The function `describe_list()` produced a clean summary of `sample_data`. This kind of reusable function is useful when you need to inspect multiple datasets quickly without rewriting the same print statements.

## Section 5: Common Mistakes in Markdown

When writing Markdown in Jupyter Notebook, beginners often make the following mistakes:

### Mistakes to Avoid

1. **Forgetting the space after `#`** — Writing `#Heading` instead of `# Heading` will not render the heading correctly.
2. **Wrong cell type** — Writing Markdown syntax in a Code cell will cause errors or produce no output. Always change the cell type to Markdown first.
3. **Using single quotes instead of backticks** — Inline code uses backticks (`` ` ``), not single quotes (`'`). The backtick key is at the top-left of most keyboards.
4. **Not running the cell** — A Markdown cell must be executed with `Shift + Enter` to render the formatted output.
5. **Incorrect indentation for nested lists** — Sub-items must be indented with at least two spaces, not a tab.

### Quick Reference: Markdown Syntax

| Element | Syntax | Example |
|---|---|---|
| H1 Heading | `# Text` | Project Title |
| H2 Heading | `## Text` | Section Name |
| H3 Heading | `### Text` | Subsection |
| Unordered list | `- Item` | Bullet point |
| Ordered list | `1. Item` | Numbered step |
| Inline code | `` `code` `` | Variable name |
| Code block | ` ```python ``` ` | Multi-line snippet |
| Bold | `**text**` | Emphasis |
| Italic | `*text*` | Light emphasis |
