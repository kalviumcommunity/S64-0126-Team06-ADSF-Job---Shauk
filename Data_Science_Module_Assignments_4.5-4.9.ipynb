# Assignment 4.9: Running, Restarting, and Interrupting Jupyter Kernels

## Objective
This assignment explains how the Jupyter kernel works, how to execute code cells, and how to manage the kernel through restarting and interrupting operations. Understanding kernel management is crucial for debugging issues and maintaining a clean computational environment.

## What is a Kernel?

### Definition
A **kernel** is a separate Python process that executes code and maintains the state of variables in memory. When you run a Jupyter notebook, the kernel is what actually interprets and executes your Python code.

### Why This Matters
- **State persistence**: Variables defined in one cell are available to all subsequent cells
- **Resource management**: The kernel uses system memory to store variables and data
- **Isolation**: Each notebook has its own kernel (or can share one deliberately)
- **Performance**: A "stale" kernel with old variables can slow down your work

### Visual Representation
```
Jupyter Notebook Interface (Browser)
              ↓
    [You write Python code in cells]
              ↓
    Sends code to Kernel
              ↓
    Kernel (Python Process)
    ├── Executes code
    ├── Stores variables in memory
    ├── Returns output/errors
    └── Maintains state
              ↓
    Returns result to Browser
              ↓
    [You see output in notebook]
```

## Understanding Execution Status

### In [ ]: (Empty Brackets)
- Cell has never been run
- Or kernel has been restarted

### In [*]: (Asterisk)
- Cell is **currently running**
- Don't run another cell until this completes
- Most common with long-running operations

### In [3]: (Number)
- Cell has been executed
- Number indicates **execution order**, not line number
- Example: `In [3]` means this was the 3rd cell executed

### Out [3]: (Output)
- Result of the cell execution
- Appears below the code cell
- Only appears if code produces output

## Example: Execution Status in Action

**Scenario: Running cells out of order**

```
Cell 1:
In []:    x = 10

Cell 2:
In []:    y = 20

Cell 3:
In []:    print(x + y)
```

**Execution sequence:**
1. You run Cell 3 first
   - Error: `NameError: name 'x' is not defined`
   - Cell status: `In [1]:`

2. You run Cell 1
   - No output (assignment operation)
   - Cell status: `In [2]:`

3. You run Cell 2
   - No output
   - Cell status: `In [3]:`

4. You run Cell 3 again
   - Output: `30`
   - Cell status: `In [4]:` (new number because it's the 4th execution)

**Key insight**: Variables `x` and `y` must be defined (run) before you use them.

## Restarting the Kernel

### Why Restart?

**Reasons to restart a kernel:**

1. **Clean slate**: Remove all variables from memory to start fresh
2. **Fix contaminated state**: You modified a variable incorrectly and can't remember the original value
3. **Resolve import errors**: You updated a library and need to reload it
4. **Memory management**: Release memory used by large variables you no longer need
5. **Remove crashed state**: If something went wrong, restarting can recover functionality

### What Happens When You Restart?

- **Cleared**: All variables are deleted
- **Preserved**: Your code in cells is NOT deleted (it's still visible)
- **Reset**: Execution counter resets to empty brackets `In [ ]:`
- **Effect**: Next time you run a cell, it's like starting fresh

### Method 1: Restart Using Menu

1. In the menu: **Kernel > Restart**
2. A dialog box appears asking to confirm
3. Click **"Restart"** to confirm
4. Kernel is restarted; all variables are cleared

### Method 2: Restart and Run All Cells

1. In the menu: **Kernel > Restart & Run All**
2. Kernel restarts AND automatically runs all cells from top to bottom
3. Useful to ensure entire notebook works from clean state
4. Warning: This takes time with many cells

### Method 3: Keyboard Shortcut

**Press `0` twice quickly** to restart kernel
- First press "0", release, press "0" again
- Actually only `Ctrl+Shift+R` reliably restarts (varies by system)

### Example Scenario: When to Restart

**Problem**: You're analyzing data and accidentally run:
```python
# Cell 5:
df = None  # Oops! You meant to delete a different variable
```

Now `df` is None, and all subsequent cells fail. Options:

**Option A: Fix the mistake**
1. Edit Cell 5 to restore `df` correctly
2. Run Cell 5 again

**Option B: Restart and re-run**
1. Kernel > Restart
2. Cell > Run All
3. Gets you back to working state automatically

## Interrupting Code Execution

### Why Interrupt?

Code sometimes runs longer than expected:
- **Infinite loop**: Cell runs forever
- **Slow operation**: Data processing takes hours unexpectedly
- **Mistake**: You realize you made an error while it's running

When you see `In [*]:`, the cell is still running. If it's taking too long, you can interrupt it.

### Method 1: Click the Stop Button

In the toolbar, click the **Stop button (■ square)** to interrupt execution.

### Method 2: Keyboard Shortcut

**Press `Ctrl+C`** (Windows/Linux) or **`Cmd+C`** (macOS)

- Sends interrupt signal to kernel
- Stops execution of current cell
- Any code that ran before the interrupt is still in memory

### Example: Interrupting

**Cell that will run a long time:**
```python
import time

print("Starting long operation...")
for i in range(1000000):
    time.sleep(0.01)  # Sleep 10 milliseconds per iteration
    if i % 100 == 0:
        print(f"Progress: {i}")

print("Done!")
```

**If you interrupt after 15 seconds:**
- Kernel stops the loop
- Partial output is visible (printed values up to interruption)
- Cell status shows it was interrupted
- Variables previously defined remain available

## Troubleshooting Common Kernel Issues

### Issue: Cell Runs Forever (Infinite Loop)

**Symptom**: `In [*]:` never changes, cell never finishes

**Solution**:
1. Click the Stop button or press `Ctrl+C`
2. Fix the loop condition in your code
3. Run cell again

**Example fix:**
```python
# Before (infinite loop):
i = 0
while True:
    print(i)
    # Forgot to increment i!

# After (fixed):
i = 0
while i < 10:
    print(i)
    i += 1  # Now loop terminates
```

### Issue: NameError - Variable Not Defined

**Symptom**: Error `NameError: name 'variable_name' is not defined`

**Cause 1**: Cell defining variable wasn't run yet
**Solution**: Run the cell that defines the variable first

**Cause 2**: Kernel was restarted and variable exists only in memory
**Solution**: Run the defining cell again, or restart and run all cells

**Cause 3**: Variable name is misspelled
**Solution**: Check spelling and capitalization

### Issue: Module Not Found Error

**Symptom**: `ModuleNotFoundError: No module named 'pandas'`

**Cause**: Library wasn't imported or isn't installed

**Solution**:
```python
# In terminal (not notebook):
conda install pandas

# OR in notebook cell:
import subprocess
subprocess.check_call(['conda', 'install', '-y', 'pandas'])

# Then restart kernel
```

### Issue: Memory Running Out

**Symptom**: Cell runs slowly or fails, messages about memory

**Cause**: Large dataframes or variables using too much RAM

**Solution**:
```python
# Delete large variables no longer needed:
del df1, df2, large_array

# Restart kernel to free all memory:
# Kernel > Restart

# Process data in chunks rather than all at once
```

# Assignment 4.8: Understanding Notebook Cells - Code vs Markdown

## Objective
This assignment explains the two fundamental cell types in Jupyter Notebook: Code cells and Markdown cells. Understanding the difference between them and knowing when to use each is essential for creating well-organized, reproducible data science work.

## Why Mix Code and Markdown?
- **Reproducibility**: Document your reasoning alongside the code that implements it
- **Communication**: Explain to others (and to future you) why you made certain choices
- **Learning**: Reinforce understanding by articulating problems and solutions
- **Professionalism**: Present polished analysis with proper structure and explanation

## The Two Cell Types

### Cell Type 1: Code Cells

**Purpose**: Execute Python (or other language) code

**Characteristics**:
- Contains executable Python code
- Preceded by `In [ ]` on the left margin (empty) or `In [n]` (executed)
- Displays output below the cell (if any)
- Can be run independently
- Execution order matters (variables carry over between cells)

**Visual indicator**: Gray border/background, code displayed in monospace font

**Example Code Cell Structure**:
```
In [ ]:
│ x = 5
│ y = 10
│ z = x + y
│ print(f"The sum is: {z}")
```

When executed, displays:
```
The sum is: 15
```

---

### Cell Type 2: Markdown Cells

**Purpose**: Display formatted text, explanations, and documentation

**Characteristics**:
- Contains text formatted with Markdown syntax
- Does not execute code
- Rendered as formatted text rather than plain code
- Supports headers, lists, links, images, and mathematical equations
- Helps organize notebook structure

**Visual indicator**: Plain text rendering, no `In [ ]` marker

**Example Markdown Cell**:
```markdown
## Data Analysis Section

This section analyzes **customer data** from Q1 2024.

Key findings:
- Sales increased by 15%
- Customer retention improved to 92%
- New markets opened in 3 regions
```

When rendered, displays:
```
Data Analysis Section
This section analyzes customer data from Q1 2024.

Key findings:
- Sales increased by 15%
- Customer retention improved to 92%
- New markets opened in 3 regions
```

## Cell Execution Order

### Important Concept: Kernel Memory

All variables defined in code cells are stored in the kernel's memory and accessible to all subsequent cells, **regardless of their position in the notebook**.

### Example:

### Best Practice:
Always run cells from top to bottom to ensure variables are defined before use.

## Step-by-Step Installation Guide

### For macOS

**Step 1: Download Anaconda**
1. Visit the official Anaconda website: https://www.anaconda.com/download
2. Select the macOS version
3. Choose the appropriate architecture (Intel or Apple Silicon/M1)
4. Click "Download" (the file will be approximately 700 MB)

**Step 2: Run the Installer**
1. Open the Downloads folder
2. Double-click the file `Anaconda3-...-MacOSX-x86_64.sh` or `.pkg` file
3. Follow the installer instructions by clicking "Continue" through the setup wizard
4. Accept the license agreement
5. Choose "Install for me only" when prompted (recommended for single-user setup)
6. Choose the default installation location: `/Users/YOUR_USERNAME/opt/anaconda3`

**Step 3: Initialize Conda**
1. Open Terminal (Applications > Utilities > Terminal)
2. Run the following command:
   ```bash
   conda init zsh
   ```
   (Use `bash` instead of `zsh` if your terminal is bash)
3. Restart your terminal or run: `source ~/.zprofile`

---

4. Accept the license terms

**Step 3: Verify Installation**
1. Open Command Prompt or PowerShell
